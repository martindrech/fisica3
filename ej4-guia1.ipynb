```python
import numpy as np
import matplotlib.pyplot as plt

def E_point(x, y, q, x0, y0):
    """
    Da el campo de una carga puntual q, ubicada en (x0, y0) en el punto (x, y)
    """
    r = np.sqrt((x-x0)**2+(y-y0)**2)
    Ex = q*(x-x0)/r**3
    Ey = q*(y-y0)/r**3
    return Ex, Ey    



def multiple_fields(x, y, charges_positions):
    """
    Da el campo en un punto (x, y) de muchas cargas. 
    
    charges_positions: las cargas que quiero agregar (q, x0, y0).
    """
    Exs, Eys = [], []
    for data in charges_positions:
        q, x0, y0 = data
        Ex, Ey = E_point(x, y, q, x0, y0)
        Exs.append(Ex)
        Eys.append(Ey)
    
    Ex_total = np.sum(Exs, 0)
    Ey_total = np.sum(Eys, 0)
    
    return Ex_total, Ey_total
    
def plot_charges(charges_positions, ax):
    """
    Simplemente grafica las cargas como pelotitas. 
    """
    for data in charges_positions:
        q, x0, y0 = data
        c = 'r'
        si = '-'
        if q > 0:
            c = 'b'
            si = '+'
        ax.plot(x0, y0, 'o'+c, markersize = 20)
```

```python
l = 3 # tamaño de la grilla
N = 16# numero de puntos de la grilla 

X, Y = np.meshgrid(np.linspace(-l, l, N), 
                   np.linspace(-l, l, N))

#una carga puntual
charges_positions = [[1, 0, 0]]



Ex, Ey = multiple_fields(X, Y, charges_positions)
n = np.sqrt(Ex**2+Ey**2)

plt.figure(3, figsize = (6,6))
fig, ax = plt.subplots(1, 1, num = 3)

plt.quiver(X, Y, Ex/n, Ey/n)
                
plot_charges(charges_positions, ax)

# Setting x, y boundary limits
ax.grid('On')
ax.set_xlim(-l, l)
ax.set_ylim(-l, l)
ax.set_aspect('equal')
ax.set_title('Campo normalizado de una carga puntual positiva')
plt.tight_layout()
```
